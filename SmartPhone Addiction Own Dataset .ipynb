```python
from sklearn.preprocessing import LabelEncoder

# List of Categorical Columns
categorical_columns = ['What is your occupation? ',
                       'How many hours do you spend on your smartphone daily?', 
                       'What is the primary purpose of your smartphone usage?', 
                       'Which apps do you use the most?',
                       'Do you feel anxious or restless when you are unable to use your smartphone?',
                       'Have you ever tried to reduce your smartphone usage but failed?',
                       'Do you prioritize smartphone activities over face-to-face interactions?',
                       'Has your smartphone usage negatively affected your sleep?',
                       'Do you often use your smartphone during meals or family gatherings?',
                       'Have you noticed physical discomfort (e.g., eye strain, neck pain) due to prolonged smartphone use?',
                       'Do you frequently check your phone without any specific reason?',
                       'Have you ever ignored important tasks to spend more time on your smartphone?',
                       'Do you set specific limits on your smartphone usage?',
                       'Do you use your smartphone to cope with stress or loneliness?',
                       'Do you feel satisfied with your current smartphone usage habits?']

# Initialize Label Encoder
label_encoder = LabelEncoder()

# Apply Label Encoding
for col in categorical_columns:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = label_encoder.fit_transform(df[col])


```

```python
import numpy as np
import re  # Import regex module for cleaning names

# Function to clean filenames (removes ?, :, *, etc.)
def clean_filename(name):
    return re.sub(r'[<>:"/\\|?*]', '', name).strip().replace(' ', '_')

# Save LabelEncoders for categorical columns
for col in categorical_columns:
    filename = f"{clean_filename(col)}_classes.npy"
    np.save(filename, label_encoder.classes_)

print("Label encoder mappings saved successfully!")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier

# SMOTE Oversampling to Fix Imbalanced Data
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size=0.2, random_state=42)

# List of Models
models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(kernel='linear', random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}

# Loop through each model
for name, model in models.items():
    print("="*50)
    print(f"Training {name} Model 🔥")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"\n{name} Results:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("="*50)
```

```python
def predict_input():
    input_data = []
    print("Enter the following details:")

    # Gender
    #gender = input("What is your gender? (Male=1, Female=0): ")
    #input_data.append(float(gender))

    #Occupation
    occupation = input("Occupation (Student=0, Professional=1, Self-employed=2, Unemployed=3, Other=4): ")
    input_data.append(float(occupation))

    # Screen Time
    screen_time = input("How many hours do you spend on your smartphone daily? (Less than 1 hour=0, 1–3 hours=1, 3–5 hours=2, More than 5 hours=3): ")
    input_data.append(float(screen_time))

    # Primary Purpose
    purpose = input("Primary Purpose (Social media=0, Work=1, Entertainment=2, Communication=3, Other=4): ")
    input_data.append(float(purpose))

    # Most Used Apps
    apps = input("Which apps do you use the most? (Social media=0, Gaming=1, Messaging=2, Productivity=3, Streaming=4, Other=5): ")
    input_data.append(float(apps))

    # Behavioral Factors (Yes=1, No=0)
    questions = [
        "Do you feel anxious or restless when you are unable to use your smartphone?",
        "Have you ever tried to reduce your smartphone usage but failed?",
        "Do you prioritize smartphone activities over face-to-face interactions?",
        "Has your smartphone usage negatively affected your sleep?",
        "Do you often use your smartphone during meals or family gatherings?",
        "Have you noticed physical discomfort (e.g., eye strain, neck pain) due to prolonged smartphone use?",
        "Do you frequently check your phone without any specific reason?",
        "Have you ever ignored important tasks to spend more time on your smartphone?",
        "Do you set specific limits on your smartphone usage?",
        "Do you use your smartphone to cope with stress or loneliness?",
        "Do you feel satisfied with your current smartphone usage habits?"
    ]

    for question in questions:
        response = input(f"{question} (Yes=1, No=0): ")
        input_data.append(float(response))

    # Convert input into array and predict
    input_data = [input_data]
    prediction = rf_model.predict(input_data)

    result = "Low" if prediction == 0 else "Moderate" if prediction == 1 else "High"
    print("\nPredicted Addiction Level:", result)

predict_input()
```
